**Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).**

**1. Artificial Intelligence (AI)**

Definition: AI is the broad field of computer science that focuses on building systems capable of performing tasks that normally require human intelligence.

Scope: Includes reasoning, learning, problem-solving, perception, natural language understanding, and decision-making.

Example: A chatbot that understands and responds to customer queries.

**2. Machine Learning (ML)**

Definition: ML is a subset of AI that allows systems to learn from data without being explicitly programmed.

How it works: Uses algorithms to find patterns in data and make predictions or decisions.

Example: An email spam filter that learns from past emails to classify new ones.

**3. Deep Learning (DL)**

Definition: DL is a subset of ML that uses artificial neural networks with many layers (hence “deep”).

Strength: Works very well with large datasets and unstructured data like images, audio, and text.

Example: Self-driving cars using image recognition to detect pedestrians and traffic signals.

**4. Data Science (DS)**

Definition: DS is the field that combines statistics, data analysis, and ML to extract insights and knowledge from data.

Scope: Covers the whole data pipeline → collecting, cleaning, analyzing, visualizing, and building predictive models.

Example: An e-commerce company analyzing purchase data to recommend products and forecast sales. 

**Question 2: What are the types of machine learning? Describe each with one
real-world example.**

Types of Machine Learning

Machine Learning is generally categorized into three main types (plus a newer emerging fourth one).

**1. Supervised Learning**

Definition: The model is trained on a labeled dataset (input → output is known).

Goal: Learn a mapping between inputs and outputs to predict results for new unseen data.

Techniques: Regression, Classification.

Example:

Spam Email Detection → Emails (input) are labeled as spam or not spam (output).

The model learns from these labels and can classify new emails correctly.

**2. Unsupervised Learning**

Definition: The model is trained on unlabeled data (only inputs are given, no outputs).

Goal: Discover hidden patterns, groupings, or structures in data.

Techniques: Clustering, Dimensionality Reduction.

Example:

Customer Segmentation → An e-commerce company groups customers into clusters (e.g., frequent buyers, discount-seekers, one-time buyers) without predefined labels.

**3. Reinforcement Learning (RL)**

Definition: The model (called an agent) learns by interacting with an environment and receiving rewards or penalties as feedback.

Goal: Maximize long-term rewards by learning the best sequence of actions.

Example:

Self-Driving Cars → The car (agent) receives rewards for safe driving (staying in lane, avoiding collisions) and penalties for mistakes (crashes, traffic violations).

**4. Semi-Supervised Learning (Hybrid)**

Definition: Uses a mix of small labeled data + large unlabeled data.

Goal: Improve learning accuracy when labeled data is scarce.

Example:

Medical Diagnosis → Only a few X-rays are labeled by doctors, but a large set is unlabeled. The model uses both to improve disease detection.

**Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.**

**1. Overfitting**

Definition: Overfitting happens when a model learns the training data too well, including noise and random fluctuations, instead of just the underlying patterns.

Effect:

Very high accuracy on training data

Poor accuracy on unseen/test data (low generalization)

Example: A decision tree that memorizes every training example instead of learning general rules.

**2. Underfitting**

Definition: Underfitting occurs when a model is too simple to capture the underlying structure of the data.

Effect:

Poor performance on both training and test data.

Example: Trying to fit a straight line (linear regression) to highly non-linear data.

**3. Bias-Variance Tradeoff**

Bias: Error due to over-simplified assumptions in the model. High bias = underfitting.

Variance: Error due to too much sensitivity to training data (model memorizes noise). High variance = overfitting.

Tradeoff:

Increasing model complexity → lowers bias but increases variance.

Decreasing model complexity → lowers variance but increases bias.

Goal: Find the sweet spot (optimal model complexity) where both bias and variance are balanced, giving good generalization.

**Question 4: What are outliers in a dataset, and list three common techniques for
handling them.**

**Outliers in a Dataset**

Definition: Outliers are data points that significantly differ from the rest of the observations in a dataset.

Impact:

Can distort statistical measures (mean, variance, correlation).

Can negatively affect machine learning models, especially ones sensitive to scale (e.g., linear regression, k-means).

Example: In exam scores [50, 52, 55, 57, 95], the score 95 may be considered an outlier.

**Techniques for Handling Outliers**
**1. Removal (Deleting Outliers)**

Remove extreme values if they are due to errors, noise, or irrelevant cases.

Example: If age in a dataset is recorded as 250 years, it’s likely a data entry error → remove it.

**2. Transformation (Reducing Impact)**

Apply mathematical transformations to reduce the effect of outliers.

Common techniques:

Log Transformation

Square Root Transformation

Example: Taking log of highly skewed income data reduces the gap between very high earners and others.

**3. Imputation / Capping (Replacing or Limiting)**

Winsorization (Capping): Replace extreme values with a threshold value (e.g., replace values above 95th percentile with the 95th percentile).

Imputation: Replace outliers with mean, median, or predicted values.

Example: In salary data, replace extremely high salaries (outliers) with the 95th percentile salary.

**Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.**

**Handling Missing Values**
**1. Detect Missing Values**

Check for missing data using tools like .isnull() or .info() in Pandas.

Identify whether the data is MCAR (Missing Completely At Random), MAR (Missing At Random), or MNAR (Missing Not At Random).

**2. Common Ways to Handle Missing Data**

Remove Missing Values

Drop rows or columns with too many missing values.

Works if missing percentage is very low.

Impute Missing Values

Fill in missing data with statistical values or predicted values.

Ensures no data loss.

**3. Use Advanced Methods**

Machine learning models (like KNN imputer, regression imputation) to predict missing values.

Imputation Techniques
For Numerical Data → Mean/Median Imputation

Mean Imputation: Replace missing values with the mean of that column.

Median Imputation: Better for skewed data (less affected by outliers).

Example: If Age = [25, 30, NaN, 40] → Replace NaN with mean (31.6) or median (30).

**For Categorical Data → Mode Imputation**

Mode Imputation: Replace missing categorical values with the most frequent category.

Example: If Color = [Red, Blue, NaN, Red] → Replace NaN with "Red" (most frequent).

**Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.
(Include your Python code and output in the code box below.)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Convert to DataFrame
df = pd.DataFrame({"Values": data})

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, "Values"] = np.nan

print("Number of missing values:", df["Values"].isnull().sum())

# 3. Fill missing values with column mean
mean_value = df["Values"].mean()
df_filled = df.copy()
df_filled["Values"].fillna(mean_value, inplace=True)

# 4. Plot histogram before and after imputation
plt.figure(figsize=(12, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(df["Values"].dropna(), bins=30, edgecolor="black")
plt.title("Before Imputation (Missing Values Removed)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(df_filled["Values"], bins=30, edgecolor="black", color="orange")
plt.title("After Imputation (Mean Filled)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

**Question 10: You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:**

**● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.
(Include your Python code and output in the code box below.)**

**ANSWER**
**1.Handle Missing Values (e.g., Ages)**

For numerical columns (like Age):

Option 1: Fill with mean/median.

Option 2: Use more advanced imputation (e.g., KNNImputer).

For categorical columns: Fill with mode (most frequent).

**2.Detect and Handle Outliers (Transaction Amounts)**

Use IQR (Interquartile Range) or Z-score.

Option: Cap (winsorize) or remove outliers.

**3. Handle Class Imbalance (Fraud vs. Non-Fraud)**

Use resampling techniques:

SMOTE (oversample minority class).

RandomUnderSampler (undersample majority class).

Or adjust class weights in the model.

**4. Encode Categorical Variables (e.g., Payment Method)**

If nominal (no order): One-Hot Encoding.

If ordinal (has order): Label Encoding.

**5. Feature Scaling**

Apply Min-Max or StandardScaler if needed for distance-based algorithms (e.g., KNN, SVM).

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from imblearn.over_sampling import SMOTE

# ------------------------------
# Step 1: Create synthetic dataset
# ------------------------------
data = {
    "Age": [25, np.nan, 40, 35, np.nan, 60, 22],
    "TransactionAmount": [100, 2000, 150, 300, 250000, 180, 90],
    "PaymentMethod": ["Credit", "Debit", "Cash", "Credit", "Debit", "Wallet", "Cash"],
    "Fraud": [0, 0, 0, 0, 1, 0, 1]  # Imbalanced target
}
df = pd.DataFrame(data)
print("Original Data:\n", df)

# ------------------------------
# Step 2: Handle Missing Values (Age)
# ------------------------------
imputer = SimpleImputer(strategy="median")
df["Age"] = imputer.fit_transform(df[["Age"]])

# ------------------------------
# Step 3: Handle Outliers (TransactionAmount)
# Using IQR method
# ------------------------------
Q1 = df["TransactionAmount"].quantile(0.25)
Q3 = df["TransactionAmount"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers instead of removing
df["TransactionAmount"] = np.where(df["TransactionAmount"] > upper_bound, upper_bound,
                                   np.where(df["TransactionAmount"] < lower_bound, lower_bound,
                                            df["TransactionAmount"]))

# ------------------------------
# Step 4: Encode Categorical Variables (PaymentMethod)
# ------------------------------
encoder = OneHotEncoder(sparse_output=False, drop="first")  # drop first to avoid dummy trap
encoded = encoder.fit_transform(df[["PaymentMethod"]])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["PaymentMethod"]))
df = pd.concat([df.drop("PaymentMethod", axis=1), encoded_df], axis=1)

# ------------------------------
# Step 5: Handle Imbalance (Fraud target)
# ------------------------------
X = df.drop("Fraud", axis=1)
y = df["Fraud"]

smote = SMOTE(random_state=42, k_neighbors=1)
X_res, y_res = smote.fit_resample(X, y)

print("\nAfter Preprocessing:\n", X_res.head())
print("\nTarget Distribution After SMOTE:\n", y_res.value_counts())
```
